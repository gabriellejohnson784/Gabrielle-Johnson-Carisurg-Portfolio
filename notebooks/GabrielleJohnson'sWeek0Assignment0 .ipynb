```python
file_path='/content/drive/MyDrive/EmergencyTriageDataset_Reduced_Dirty.csv'
df_raw=pd.read_csv(file_path)
df_raw.head(10)
```

```python
print(df_raw.dtypes) #viewing the data type
```

```python
print(df_raw['Gender'].unique()) #viewing all the unique values in the gender column
print(f"\nTotal unique values: {df_raw['Gender'].nunique()}")
```

```python
print(df_raw['Gender'].value_counts())
```

```python
df_raw['Gender']=(df_raw['Gender'].astype(str).str.strip().str.lower()) #stripping all strings to lowercase
print(df_raw['Gender'].value_counts())
```

```python
gender_map={
    'female': '0',
    '0': '0',
    'male': '1',
    '1': '1'
}
df_raw['Gender_Clean'] = df_raw['Gender'].map(gender_map).fillna('Unknown') #mapping existing values in the column to 1/0 or Unknown if any values outside of what was defined is inputted
print(df_raw['Gender_Clean'].value_counts())
```

```python
gender_one_hot=pd.get_dummies(df_raw['Gender_Clean'], prefix='Gender',dtype=int) #use one hot encoding to help with scalabilty and flexibility of the varying genders that may be inputted
df_raw=pd.concat([df_raw, gender_one_hot], axis=1)
print(gender_one_hot.head(10))
```

```python
df_raw = df_raw.drop(columns=['Gender', 'Gender_Clean'])
```
